Cross Validating the models to select the model

I have taken the below algorithms because our problem is categorized into a classification model. To validate our dataset I have taken these classical machine learning algorithms to validate the data.

Model 1 : Support Vector Machines

```python
from sklearn.model_selection import cross_val_score # importing the cross validation function
from sklearn import svm # importing svm modue
clf = svm.SVC(kernel='linear', C=1, random_state=42) # creating the object for training
scores = cross_val_score(clf, X, Y, cv=5) # appllying cross vallidation to get performance of the model on different sets of data

scores # crossvalidated scores
```

```python
scores.mean() # mean of scores
```

Model 2 Random Forest Classifier

```python
from sklearn.model_selection import cross_val_score  # importing the cross validation function
from sklearn.ensemble import RandomForestClassifier # importing module
clf = clf = RandomForestClassifier(max_depth=2, random_state=0) # creating moe
scores = cross_val_score(clf, X, Y, cv=5) # cross validation function

scores 
```

```python
scores.mean()
```

Model 3 Decision Tree Classifier

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
scores = cross_val_score(clf, X, Y, cv=5)

scores 
```

# Phase 3 

Implementing KNN Classifier on the Dataset

```python
from sklearn.model_selection import train_test_split # importing function train_test_split
```

```python
Train_X,Test_X,Train_Y,Test_Y=train_test_split(X,Y,test_size=0.4,shuffle=True) # splitting the data into training and testing in 60 & 40 ratio
```

Training KNN Classiifer ( Selected model)

```python
from sklearn.neighbors import KNeighborsClassifier # importing the KNN classifier function
model= KNeighborsClassifier(n_neighbors=1,leaf_size=15, p=1, metric='minkowski', metric_params=None, n_jobs=10) # KNN model object creation and hyperparameter tuning
model.fit(Train_X,Train_Y) # training the KNN model on training data (X,Y)
```

```python
y_pred=model.predict(Test_X) # making predictions with Test dataset 40
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report # importing performance metris to analsyze the performance of selected model on test dataset
```

```python
cm=confusion_matrix(y_pred,Test_Y) # calculating the confusion matrrix
```

```python
cm # printing the confusion matrix
```

Confusion Matrix Visualization

Confusion Matrix is a performance measurement for machine learning classification problem where output can be two or more classes. It is a table with 4 different combinations of predicted and actual values.

We describe predicted values as Positive and Negative and actual values as True and False.

True Positive
True Negative
False Positive: (Type 1 Error)
False Negative: (Type 2 Error)




Out of topic example:

To understand TP, FP, FN, TN in terms of pregnancy analogy.




#True Positive:
Interpretation: You predicted positive and it’s true.
You predicted that a woman is pregnant and she actually is.

#True Negative:
Interpretation: You predicted negative and it’s true.
You predicted that a man is not pregnant and he actually is not.

#False Positive: (Type 1 Error)
Interpretation: You predicted positive and it’s false.
You predicted that a man is pregnant but he actually is not.

#False Negative: (Type 2 Error)
Interpretation: You predicted negative and it’s false.
You predicted that a woman is not pregnant but she actually is.
